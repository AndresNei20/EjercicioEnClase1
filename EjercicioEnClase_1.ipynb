```python
# prompt: how can i import the csv that its in the same folder

import pandas as pd

# Assuming the CSV file is named 'Latest_Data_Science_Salaries.csv' and is in the same folder as your notebook.
df = pd.read_csv('Latest_Data_Science_Salaries.csv')
df.head(10)
```

```python
# Calculate descriptive statistics for relevant variables
relevant_variables = ['Salary', 'Salary in USD']
df[relevant_variables].describe()
```

```python
# Median of Salaries (Salary and Salary in USD)
median_salary = df['Salary'].median()
median_Salary_in_USD = df['Salary in USD'].median()
print(f"Median Salary: {median_salary}")
print(f"Median Salary in USD: {median_Salary_in_USD}")
```

# Acciones por ejecutar

**1. ¿Cúantos posibles "Job Title" existen en el Dataset?**

```python
df_job_title = df['Job Title']
allJobs_df_job_title = df_job_title.unique()
count_df_job_title = df_job_title.nunique()
count_each_df_job_title = df_job_title.value_counts()
print(f"Number of unique Job Titles: {count_df_job_title}")
print(f"Unique Job Titles: {allJobs_df_job_title}")
print(f"Count of each Job Title:\n{count_each_df_job_title}")
```

**2. ¿Cúales son los "Employment Type" y en qué frecuencia se presentan en el Dataset?**

```python
df_employment_type = df['Employment Type']
count_df_employment_type = df_employment_type.value_counts()
print(f"Count of each Employment Type:\n{count_df_employment_type}")
```

**3. ¿Cuáles son los "Expertise Level" y "Experience Level" y cuál es el salario promedio general de cada uno?**

```python
df_expertise_level = df['Expertise Level']
df_experience_level = df['Experience Level']
print(f"Unique Expertise Levels:\n{df_expertise_level}")
print(f"Unique Experience Levels:\n{df_experience_level}")

df.groupby(['Expertise Level', 'Experience Level'])['Salary'].mean().round(2)
```

**4. ¿Cuáles son los "Salary Currency" y a qué tasa está covertido respecto al dolar cada uno?**

```python
salary_currencies = df['Salary Currency'].unique()
conversion_rates = {}

for currency in salary_currencies:

    currency_data = df[df['Salary Currency'] == currency]

    conversion_rate = (currency_data['Salary'] / currency_data['Salary in USD']).mean()

    conversion_rates[currency] = conversion_rate


print("Tasas de conversión respecto al USD:")
print(conversion_rates)
```

**5. ¿En cuántos paises diferentes se encuentran los empleados ("Employee Residence")?**

```python
residence_countries = df['Employee Residence'].unique()
count_residence_countries = len(residence_countries)
print(f"Unique Countries of Residence:\n{residence_countries}")
print(df['Employee Residence'].value_counts())
print(f"Number of different countries in Employee Residence: {count_residence_countries}")
```

# 6. ¿Cuál salario promedio por "Job Title" agrupado?

```python
salary_per_job_title = df.groupby('Job Title')['Salary in USD'].mean().round(2)
print(salary_per_job_title)
```

# 7 ¿Cuál es el salario promedio por país de residencia (Employee Residence)?

```python
salary_per_residence = df.groupby('Employee Residence')['Salary in USD'].mean().round(2)
print(salary_per_residence)
```

# 8. ¿Cuál es el "Salary in USD" máximo y mínimo?¿A qué registro pertenece?

```python
max_salary = df['Salary in USD'].max()
min_salary = df['Salary in USD'].min()

max_salary_record = df[df['Salary in USD'] == max_salary]
min_salary_record = df[df['Salary in USD'] == min_salary]

print(max_salary_record)

print("\nMínimo salario:")
print(min_salary_record)
```

# 9. ¿Cuál es el salario promedio (en USD) por rol por tamaño de compañía?

```python
salary_avg = df.groupby(['Job Title', 'Company Size'])['Salary in USD'].mean().round(2)
print(salary_avg)
```

# 10. Muestre los datos agrupados por tamaño de la compañía, país y salario promedio.

```python
grouped_data = df.groupby(['Company Size', 'Employee Residence'])['Salary in USD'].mean().round(2)
print(grouped_data)
```
